# Part 1
- [x] Read the dataset with `pandas`


```python
import pandas as pd

data = pd.read_csv("poultry_dataset.csv")
print("Dataset shape: ", data.shape)
data.head()
```

# Part 2
- [x] Statistical summary of each column except for the "Sr No" column

```python
descriptions = data.describe()
descriptions = descriptions.drop(["Sr No"], axis=1)
descriptions
```

# Part 3
- [x] Use `matplotlib` to plot curves of:
    - [X] Temperature
    - [X] Humidity
    - [X] CO2
    - [X] NH3
    - [X] Light Intensity

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(data.shape[0])
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 8))
axes[0, 0].plot(x, data.Temperature)
axes[0, 0].set_title("Temperature")

# plot
axes[0, 1].plot(x, data.Humidity)
axes[0, 1].set_title("Humidity")

# plot
axes[1, 1].plot(x, data.CO2)
axes[1, 1].set_title("CO2")

# plot
axes[1, 0].plot(x, data.NH3)
axes[1, 0].set_title("NH3")

# plot
axes[2, 0].plot(x, data["Light Intensity"])
axes[2, 0].set_title("Light Intensity")
```

# Part 4
- [x] Generate boxplots for
    - [x] Temperature
    - [x] Humidity
    - [x] CO2
    - [x] NH3
    - [x] Light Intensity

```python
fig2, axes2 = plt.subplots(nrows=3, ncols=2, figsize=(12, 8))
axes2[0, 0].boxplot((x, data.Temperature))
axes2[0, 0].set_title("Temperature")

# plot
axes2[0, 1].boxplot((x, data.Humidity))
axes2[0, 1].set_title("Humidity")

# plot
axes2[1, 1].boxplot((x, data.CO2))
axes2[1, 1].set_title("CO2")

# plot
axes2[1, 0].boxplot((x, data.NH3))
axes2[1, 0].set_title("NH3")

# plot
axes2[2, 0].boxplot((x, data["Light Intensity"]))
axes2[2, 0].set_title("Light Intensity")
```
