```python
from typing import Literal, Tuple

import matplotlib.pyplot as plt
import pandas as pd

# craete plots
def make_plots(
    data: pd.DataFrame = None,
    xtitle: str = None,
    ytitle: str = None,
    title: str = None,
    type: Literal[
        "Heatmap",
        "Hbar Plot",
        "K-Density",
        "Diag-Correlation",
        "Structured Heatmap",
        "All",
    ] = "All",
    figsize: Tuple[int, int] = None,
    display: bool = True,
    save_as: str = None,
    **kwargs
) -> None:
    _, axes = plt.subplots(**kwargs)

    # create heatmap
    sns.heatmap(data, annot=True, linewidths=0.5, ax=axes)

    # create barplot
    _, axes = plt.subplots(**kwargs)
    sns.set_color_codes("pastel")
    sns.barplot(x=xtitle, y=ytitle, data=data, label=title, color="b")

    # create conditional kernel density estimate
    sns.displot(
        data=data,
        x=xtitle,
        y=ytitle,
        kind="kde",
        height=6,
        multiple="fill",
        palette="ch:rot=-.25,hue=1,light=.75",
    )

    if display or PRINT_ALL:
        plt.show()

    if save_as:
        plt.savefig()

exit()
make_plots(data=DATA, xtitle="x-title test", ytitle="y-title test", title="title test")
```
