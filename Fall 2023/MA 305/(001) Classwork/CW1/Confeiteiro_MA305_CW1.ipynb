```python
#!/usr/bin/env python3
"""
========================================================================
Krystian Confeiteiro
Dr. Sam
Classwork 1
MA305-06DB

Purpose:
    Evaluate F(x) = a-b*(x**3)  at N equidistant points in [0,1],
    and output the pairs (x , F(x))  on the screen and in the file
    "out.txt" for plotting.
========================================================================
"""

# Get input from user:
print(' Please enter the values of a, b and N:')
a = input("a = ")
b = input("b = ")
N = input("N = ")

# convert strings to float types
a = float(a)
b = float(b)
N = int(N)

# Print on screen:
print("\nEntered constants:")
print("a = ", a)
print("b = ", b)
print("N = ", N)
print('\n #   x        F(x)')

# Open a file (out.txt) for output:
file = open("out.txt", "w")
print("Output file: ", file)


# Compute F(x) = a-b*(x**3)  and print  x   F(x)
dx = 1/(N-1)
for i in range(N):
    xi = i*dx
    Fi = a - b*xi**3
    print("{0: 0.7f} {1: 0.7f}".format(xi, Fi))
    file.write("{0: 0.7f} {1: 0.7f}\n".format(xi, Fi))

file.close()
```
